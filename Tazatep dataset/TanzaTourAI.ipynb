```python
group=train.isnull().sum().sort_values().plot(kind='barh')
plt.title("MISSING VALUES IN DATA SET")
for p in group.patches:
    z=((p.get_width()/train.shape[0])*100)
    per="{:,.0f}%".format(z)
    width=p.get_width()
    height=p.get_height()
    x=p.get_x() + width + 0.02 
    y=p.get_y() + height/2
    plt.annotate(per,(x,y))    
    
```
